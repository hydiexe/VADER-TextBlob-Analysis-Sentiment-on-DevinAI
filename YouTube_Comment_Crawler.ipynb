```python
import pandas as pd
from googleapiclient.discovery import build
```

```python
def komentar_video(video_id):
    komentar = []

    youtube = build('youtube', 'v3', developerKey=api_key)

    video_response = youtube.commentThreads().list(part='snippet', videoId=video_id).execute()

    # iterasi tanggapan video
    while video_response:
        for item in video_response['items']:

            # mengekstrak komentar
            published = item['snippet']['topLevelComment']['snippet']['publishedAt']
            user = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
            teks_komentar = item['snippet']['topLevelComment']['snippet']['textDisplay']
            jumlahLike = item['snippet']['topLevelComment']['snippet']['likeCount']

            komentar.append([published, user, teks_komentar, jumlahLike])

        if 'nextPageToken' in video_response:
            video_response = youtube.commentThreads().list(
                part='snippet',
                pageToken=video_response['nextPageToken'],
                videoId=video_id
            ).execute()
        else:
            break
    # endwhile
    return komentar
```

```python
# API (di dapat dari google developer console)
api_key = 'API-KEY-ANDA'

# ID video
# contoh url video = https://www.youtube.com/watch?v=fjHtjT7GO1c
id_video = "fjHtjT7GO1c"

komentar = komentar_video(id_video)

komentar
```

```python
df = pd.DataFrame(komentar, columns=['publishedAt', 'authorDisplayName', 'textDisplay', 'likeCount'])
df
```

```python
df.to_csv('komentar-youtube.csv', index=False)
```
